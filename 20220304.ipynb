```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([{"Gender": "Male", "HeightCm": 171, "WeightKg": 96 },
{ "Gender": "Male", "HeightCm": 161, "WeightKg": 85 },
{ "Gender": "Male", "HeightCm": 180, "WeightKg": 77 },
{ "Gender": "Female", "HeightCm": 166, "WeightKg": 62},
{"Gender": "Female", "HeightCm": 150, "WeightKg": 70},
{"Gender": "Female", "HeightCm": 167, "WeightKg": 82}])
df
```

```python
df['BMI'] = df['WeightKg']/(df['HeightCm']/100)**2
df
```

```python
df['BMICat'] = np.where(df['BMI']<=18.4,"Under weight",
np.where((df['BMI']>18.5) & (df['BMI']<=24.9),"Normal weight",
np.where((df['BMI']>25)   & (df['BMI']<=29.9),"Over weight",
np.where((df['BMI']>30)   & (df['BMI']<=34.9),"Moderately obese",
np.where((df['BMI']>35)   & (df['BMI']<=39.9),"Severely obese",
np.where((df['BMI']>=40),"Very severely obese","Normal"))))))
df['HealthRisk'] = np.where(df['BMI']<=18.4,"Malnutrition risk",
np.where((df['BMI']>18.5) & (df['BMI']<=24.9),"Low risk",
np.where((df['BMI']>25)   & (df['BMI']<=29.9),"Enhanced risk",
np.where((df['BMI']>30)   & (df['BMI']<=34.9),"Medium risk",
np.where((df['BMI']>35)   & (df['BMI']<=39.9),"High risk",
np.where((df['BMI']>=40),"Very high risk","Normal"))))))
df
```

```python
print('Total #No Of Over Weight People = ',len(df[df['BMICat']=='Over weight']))
```
